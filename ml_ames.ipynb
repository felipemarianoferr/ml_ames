### Justificativas Qualitativas (Baseadas no Dicionário de Dados) para as 11 Features Numéricas Selecionadas:

1.  **`Overall.Qual` (Qualidade Geral):**
    * Mede a qualidade geral dos materiais e acabamentos. Casas com maior qualidade percebida tendem a ser mais valorizadas.
2.  **`Gr.Liv.Area` (Área Habitável Acima do Solo):**
    * Representa o tamanho do espaço principal de convivência. Casas maiores geralmente custam mais.
3.  **`Garage.Cars` (Capacidade da Garagem em Carros):**
    * Indica o tamanho da garagem pela capacidade. Garagens são comodidades valorizadas, e maior capacidade associa-se a imóveis de maior valor.
4.  **`Garage.Area` (Área da Garagem):**
    * Mede o tamanho físico da garagem. Uma garagem maior oferece mais utilidade, tendendo a aumentar o valor do imóvel.
5.  **`Total.Bsmt.SF` (Área Total do Porão):**
    * Refere-se ao tamanho total do porão. Adiciona espaço útil (moradia, lazer, armazenamento), influenciando o valor. A existência/tamanho é um diferencial.
6.  **`Year.Built` (Ano de Construção):**
    * Data da construção original. Casas mais novas geralmente são mais caras (padrões/materiais modernos, menos manutenção).
7.  **`X1st.Flr.SF` (Área do Primeiro Andar):**
    * Mede o tamanho do andar térreo. Parte fundamental do espaço útil (áreas sociais), impactando o valor.
8.  **`TotRms.AbvGrd` (Total de Cômodos Acima do Solo):**
    * Indica o número de cômodos principais (excluindo banheiros). Mais cômodos geralmente sugerem casa maior/mais funcional, influenciando o valor.
9.  **`Lot.Frontage` (Testada do Lote):**
    * Medida da frente do lote na rua. Relaciona-se ao tamanho/proeminência do lote, acesso e apelo visual, fatores que influenciam o valor.
10. **`X2nd.Flr.SF` (Área do Segundo Andar):**
    * Mede o tamanho do segundo andar em casas de múltiplos níveis. Contribui para a área total e diferencia tipos de casas, impactando tamanho/valor.
11. **`Bedroom.AbvGr` (Número de Quartos Acima do Solo):**
    * Métrica padrão de capacidade residencial. Mais quartos geralmente indicam casa maior/mais familiar, influenciando o valor.

### Tirando outliers e removendo spikes

### Justificativa:

A remoção de outliers foi feita com base na análise visual dos histogramas e boxplots, que revelaram valores extremos e distribuições assimétricas. Alguns dados foram excluídos por apresentarem valores atípicos que poderiam distorcer a análise, enquanto outros foram removidos por não representarem o que aparentavam — por exemplo, valores zero em Garage.Area indicam ausência de garagem, não uma garagem muito pequena. Também foram filtradas categorias raras ou de baixa qualidade, como Bsmt.Qual = 'Po', por não contribuírem de forma significativa ao modelo. ### Construindo os modelos

```python
#Para montar as pipelines
def get_column_names():

    numerical_features = [
    'Overall.Qual',
    'Gr.Liv.Area',
    'Garage.Cars',
    'Garage.Area',
    'Total.Bsmt.SF',
    'Year.Built',
    'X1st.Flr.SF',
    'TotRms.AbvGrd',
    'Lot.Frontage',
    'X2nd.Flr.SF',
    'Bedroom.AbvGr'
    ]

    categorical_features = [
    'Exter.Qual',
    'Foundation',
    'Kitchen.Qual',
    'Bsmt.Qual'
    ]

    return numerical_features, categorical_features
```

```python
#Criando uns transformadores legais para o modelo fancy
import numpy as np
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin

class minMaxQuantile(BaseEstimator, TransformerMixin):

    """ 
        Class to create a transformer that will return 
        the min and max quantile of a given dataset.
    """

    def __init__(self, min_quantile, max_quantile):
        """ 
            Class constructor that defines the min and
            max quantile values.
        """
        self.min_quantile = min_quantile
        self.max_quantile = max_quantile
        self.min_values = None
        self.max_values = None
        self.feature_names = None
    
    def fit(self, X, y=None):
        """ 
            Calculate and store the min and max quantile values
            for each feature in the dataset or array X.
        """

        if isinstance(X, pd.DataFrame):
            self.feature_names = X.columns
            X = X.to_numpy()

        self.min_values = np.quantile(X, self.min_quantile, axis=0)
        self.max_values = np.quantile(X, self.max_quantile, axis=0)

        return self
    
    def transform(self, X, y=None):
        """ 
            Trasform the dataset or array X into a new dataset
            based on the min and max quantile values learned on the
            fit function.
            
        """
        if self.min_values is None or self.max_values is None:
            raise ValueError('Transformer not fitted yet.')
        
        is_dataframe = isinstance(X, pd.DataFrame)

        if is_dataframe:
            X = X.to_numpy()

        X_transformed = np.clip(X, self.min_values, self.max_values)

        if is_dataframe:
            return pd.DataFrame(X_transformed, columns=self.get_feature_names_out(self.feature_names))
        
        return X_transformed

    def get_feature_names_out(self, input_features):
        """ 
            Return the name of the features after the transformation.
        """
        return [f"{name}_clipped"for name in input_features]
    
class SimpleLog1pTransformer(BaseEstimator, TransformerMixin):
    """Aplica np.log1p (log(1+x)) a uma lista de colunas especificadas."""

    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        
        return self

    def transform(self, X, y=None):
        """Aplica log1p às colunas especificadas."""

        X_transformed = X.copy()
        for col in self.columns:
            X_transformed[col] = np.where(
            X_transformed[col] > 0,
            np.log10(X_transformed[col]),
            np.log10(X_transformed[col] + 1) 
            )

        return X_transformed

    def get_feature_names_out(self, input_features=None):
        """Retorna os nomes das features."""

        if input_features is None:
           
            if hasattr(self, 'columns'):
                 
                 return self.columns
            else:
                 raise ValueError("Input feature names are required.")
        return input_features
```

```python
from sklearn.dummy import DummyRegressor
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

# Baselines

# Dummy Regressor
def create_dummy_model():
    return DummyRegressor(strategy='mean')

# Modelo Linear Simples
def simple_linear_model(logColumns,
                        log=False, Quantile=False, minQuantile=0.05, MaxQuantile=0.95):
    """
    Cria uma pipeline para Regressão Linear Simples, com transformações
    log e quantile opcionais.
    """
    numerical_columns, categorical_columns = get_column_names()
    numerical_steps = []

    if log:
        log_transformer = SimpleLog1pTransformer(columns=logColumns)
        numerical_steps.append(('log', log_transformer))

    if Quantile:
        
        quantile_transformer = minMaxQuantile(min_quantile=minQuantile, max_quantile=MaxQuantile)
        numerical_steps.append(('quantile', quantile_transformer))

    numerical_steps.append(('imputer', SimpleImputer(strategy='mean')))
    numerical_steps.append(('scaler', StandardScaler()))

    numerical_pipeline = Pipeline(steps=numerical_steps)

    categorical_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(handle_unknown='ignore', drop='first', sparse_output=False)),
    ])


    preprocessing = ColumnTransformer(
        transformers=[
            ('numerical', numerical_pipeline, numerical_columns),
            ('categorical', categorical_pipeline, categorical_columns),
        ],
        remainder='passthrough',
    )

    model = LinearRegression()

    final_simple_model = Pipeline([
        ('preprocessing', preprocessing),
        ('model', model),
    ])

    return final_simple_model
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.preprocessing import PolynomialFeatures, StandardScaler

#modelo fancy

def create_complex_model(logColumns, Quantile = True, log = True, minQuantile = 0.05, MaxQuantile = 0.95):
    numerical_columns, categorical_columns = get_column_names()

    numerical_pipeline = Pipeline([])

    if log:
        log_transformer = SimpleLog1pTransformer(logColumns)
        numerical_pipeline.steps.append(('log', log_transformer))
    
    if Quantile:
        quantile_transformer = minMaxQuantile(minQuantile, MaxQuantile)
        numerical_pipeline.steps.append(('quantile', quantile_transformer))

    numerical_pipeline.steps.append(('imputer', SimpleImputer(strategy='mean')))
    numerical_pipeline.steps.append(('scaler', StandardScaler()))
    numerical_pipeline.steps.append(('poly', PolynomialFeatures(degree=2, include_bias=False)))

    categorical_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(handle_unknown='ignore', drop='first', sparse_output=False)),
    ])


    preprocessing = ColumnTransformer(
        transformers=[
            ('numerical', numerical_pipeline, numerical_columns),
            ('categorical', categorical_pipeline, categorical_columns),
        ],
        remainder='passthrough',
    )

    model = HistGradientBoostingRegressor(max_depth=5, max_iter=100)

    complex_model = Pipeline([
        ('preprocessing', preprocessing),
        ('model', model),
    ])

    return complex_model
```

```python
import optuna
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
import numpy as np

logColumns_list = [
    'Lot.Frontage',
    'Total.Bsmt.SF',
    'Gr.Liv.Area',
    'X2nd.Flr.SF',
    'X1st.Flr.SF',
    'Garage.Area'
]

def objective_fancy(trial):
    """
    Função objetivo para Optuna otimizar a pipeline complexa.
    """
    # Parâmetros da Pipeline (Controlando Transformações)
    use_log = trial.suggest_categorical('pipeline_use_log', [True, False])
    use_quantile = trial.suggest_categorical('pipeline_use_quantile', [True, False])
    # Otimizar os quantis se use_quantile for True
    if use_quantile:
        min_q = trial.suggest_float('quantile_min', 0.0, 0.1)

        max_q = trial.suggest_float('quantile_max', 0.9, 1.0)
    else:
        min_q = 0.0
        max_q = 1.0

    poly_degree = trial.suggest_categorical('poly_degree', [1, 2])

    # Parâmetros do HistGradientBoostingRegressor
    gbr_learning_rate = trial.suggest_float('gbr_learning_rate', 0.01, 0.3, log=True)
    gbr_max_iter = trial.suggest_int('gbr_max_iter', 100, 1000)
    gbr_max_depth = trial.suggest_int('gbr_max_depth', 3, 15)
    gbr_l2_reg = trial.suggest_float('gbr_l2_regularization', 1e-8, 10.0, log=True)
    gbr_min_samples_leaf = trial.suggest_int('gbr_min_samples_leaf', 5, 50)

    # --- 2. Criar e Configurar a Pipeline para este Trial ---
    
    # Cria a pipeline baseada nos parâmetros True/False sugeridos
    pipeline_instance = create_complex_model(
        logColumns=logColumns_list,
        log=use_log,
        Quantile=use_quantile,
        minQuantile=min_q,
        MaxQuantile=max_q
    )

    # Ajusta os parâmetros dos passos *dentro* da pipeline criada
    pipeline_instance.set_params(
        preprocessing__numerical__poly__degree=poly_degree,
        model__learning_rate=gbr_learning_rate,
        model__max_iter=gbr_max_iter,
        model__max_depth=gbr_max_depth,
        model__l2_regularization=gbr_l2_reg,
        model__min_samples_leaf=gbr_min_samples_leaf
    )

    # --- 3. Avaliar a Pipeline com Validação Cruzada ---
    scores = cross_val_score(
        pipeline_instance,
        X_train,
        y_train,
        cv=5,
        scoring='neg_root_mean_squared_error',
        n_jobs=-1,
        error_score='raise'
    )
    rmse_medio = -scores.mean()

    # Optuna tentará minimizar este valor
    return rmse_medio

def objective_dummy(trial):
    """Função objetivo para testar 'mean' vs 'median' no DummyRegressor."""

    dummy_strategy = trial.suggest_categorical('dummy_strategy', ['mean', 'median'])

    model = DummyRegressor(strategy=dummy_strategy)

    scores = cross_val_score(
        model,
        X_train, # X é ignorado pelo DummyRegressor
        y_train,
        cv=5,
        scoring='neg_root_mean_squared_error',
        n_jobs=-1,
        error_score='raise'
    )
    rmse_medio = -scores.mean()

    return rmse_medio

def objective_linear(trial):
    """
    Função objetivo para otimizar a pipeline Linear Simples
    com opções de log e quantile.
    """
    # 1. Sugerir Hiperparâmetros para o Trial
    use_log = trial.suggest_categorical('linear_use_log', [True, False])
    use_quantile = trial.suggest_categorical('linear_use_quantile', [True, False])

    min_q = 0.0
    max_q = 1.0

    if use_quantile:
        min_q = trial.suggest_float('linear_quantile_min', 0.0, 0.1)
        max_q = trial.suggest_float('linear_quantile_max', 0.9, 1.0)

    pipeline_instance = simple_linear_model(
        logColumns=logColumns_list,
        log=use_log,
        Quantile=use_quantile,
        minQuantile=min_q,
        MaxQuantile=max_q
    )

    scores = cross_val_score(
        pipeline_instance,
        X_train,
        y_train,
        cv=5,
        scoring='neg_root_mean_squared_error',
        n_jobs=-1,
        error_score='raise'
    )
    rmse_medio = -scores.mean()

    return rmse_medio
```

### Teste de hipótese wilcoxon

Basicamente queremos ver se podemos fazer nossa escolha baseado na estatistica, pois apenas baseado no RMSE, o modelo fancy é ligeiramente melhor que o linear, entao seguindo apenas essa ideia escolheriamos o fancy. Mas com nosso teste de hipótese poderemos saber melhor se essa diferença é real e não apenas fruto da variabilidade dos dados nos folds da validação cruzada, portanto assumiremos H0 como:

A Hipótese Nula (H0) do teste de Wilcoxon que fizemos é que não há diferença sistemática no desempenho (medido pelo RMSE da validação cruzada) entre os dois modelos que estão sendo comparados.

Mais formalmente: a mediana da diferença entre os scores pareados dos dois modelos é igual a zero.

```python
from scipy.stats import wilcoxon

best_model_name = 'Fancy_GBR'

scores_best = cv_scores[best_model_name]

alpha = 0.05
print(f"Comparando {best_model_name} com outros modelos (alpha={alpha}):\n")

for model_name, scores_other in cv_scores.items():
    if model_name == best_model_name:
        continue

    print(f"Teste: {best_model_name} vs {model_name}")

    
    stat, p_value = wilcoxon(scores_best, scores_other, alternative='greater', zero_method='zsplit')
    

    print(f"  p-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"  Resultado: {best_model_name} é estatisticamente MELHOR que {model_name}.")
    else:
        print(f"  Resultado: Não há evidência estatística suficiente para dizer que {best_model_name} é melhor que {model_name}.")
    
    print("-" * 20)
```

### Decisao com base no teste de hipótese e analise do RMSE

Optaremos por seguir com o modelo linear. Primeiramente grande parte das features parecem ter uma relacao linear com o target como pudemos observar nos gráficos vistos anteriormente. Apesar da ligeira diferenca observado no RMSE entre o modelo fancy e o linear, nao conseguimos afirmar (com 95% de confiança) que o GBR é verdadeiramente superior ao modelo Linear simples (otimizado com log), portanto ao analisar a complexidade entre os dois modelos e saber que o GBR estatisticamente nao oferece uma melhora significativa em relacao ao linear, preferimos o modelo mais simples pois ele é mais facil de interpretar, apresenta menor complexidade de treinamento/modelo com desempenho estatisticamente indistinguivel em relacao ao GBR.
